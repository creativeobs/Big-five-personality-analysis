# Data Visualization

### Univariate Analysis

```python
fig, ax = plt.subplots(1, 2, figsize=(16,12))
fig.suptitle('Age Distribution', y=0.92, fontsize=22)
i = 0

arrowprops = dict(arrowstyle="simple", alpha=0.8, color='r')
bbox=dict(boxstyle="round", alpha=0.7, color='black')


plot=sns.countplot(x = df['Age'], ax= ax[i])
plot.annotate(xytext=(1,60), xy=(0,48), text='most participants: 68', color='white', fontsize=20,
            arrowprops=arrowprops, bbox=bbox)

plot.annotate(xytext=(5,4), xy=(5,1), text='lowest', color='white', fontsize=20, ha='center', va='center',
            arrowprops=arrowprops, bbox=bbox)

i += 1
_, _, autotexts = ax[i].pie(x = df['Age'].value_counts().values, labels = df['Age'].value_counts().index, 
                            wedgeprops=dict(width=0.6), colors=colors, autopct="%.2f%%")

for autotext in autotexts:
    autotext.set_color('white')
    

fig.savefig('figures/Age Distribution.jpg', dpi=300)
fig.show()
```

# Insights

1. Seems like 21 years old has the most number of participants. However there is a mistyped data which is "20 years old".
2. Female is the most contributing gender in this study. While Prefer not to say is the least.
3. For the Year feature, third year has the most number of participants. 
4. For Course Feature, although it should be a categorical, the number of unique values is many. This might indicate that it is not a very good feature for the model. Additionally, it has many duplicate such as "Computational and Data Science".
5. For the GWA, we can see positively skewed data. The 3.7 GWA will affect the model's prediction. We can classify it as an outlier. Obviusly, majority of the students will fall under 1.00 - 2.00 grades.
6. For the OCEAN questions, the distribution of their data will obviously be under 1 - 5. Furthermore, computations for the OCEAN will still take place and will be up for analysis.

## Multivariate

```python
plt.title('Age vs Gender', fontsize=22)
sns.heatmap(pd.crosstab(df['Age'], df['Gender'], normalize=True), cmap=colors, annot=True)
plt.savefig('figures/Age vs Gender.jpg')
```

```python
plt.title('Age vs Year', fontsize=22)
sns.heatmap(pd.crosstab(df['Age'], df['Year'], normalize=True), cmap=colors, annot=True)
plt.savefig('figures/Age vs Year.jpg')
```

```python
plt.title('Gender vs Year', fontsize=22)
sns.heatmap(pd.crosstab(df['Gender'], df['Year'], normalize=True), cmap=colors, annot=True)
plt.savefig('figures/Gender vs Year.jpg')
```

```python
fig, ax = plt.subplots(1, 3, figsize=(16,8))
i = 0

for col in ['Year', 'Gender', 'Age']:
    sns.swarmplot(data=df, x=col, y='Last Semester GWA (Generated Weighted Average)?', ax=ax[i])
    i += 1
    
fig.suptitle('Demographics vs GWA', fontsize=22)
fig.tight_layout()
fig.savefig('figures/Demographics vs GWA.jpg', dpi=300)
sns.despine()
```

2. Most amount of participants(52%) are 20 to 21 years old and is female.
3. 47% female and third year & 30% male and third year.
4. There seems to be a fair amount of third years that have 1.75 and 1.5.
